# 1. 데이터 잘라내기

```python
regular = pd.read_csv("Regular_Season_batter.csv")
```

1 .ops가 0.9를 넘는 row

2 . batter_name 이 박석민인 row

```python
regular.loc[regular['OPS']>0.9]
```

```python
regular.loc[regular['batter_name']=='박석민']
```

```python
regular.loc[regular['year']>2010]
```

```python
regular.loc[regular['batter_name']=='박석민'][['year','OPS']]
```

```python
regular.loc[(regular['batter_name']=='박석민')&(regular['year']==2009)]
```

원하는 타자의 특정 연도 성적을 뽑아내었다.

# 2. for문

```python
List = ['박석민','채태인','최형우','박해민']
for name in List:
    print(regular.loc[regular['batter_name']==name]['OPS'].mean())
```

for문을 이용하여 원하는 타자의 OPS평균을 추출하였다
